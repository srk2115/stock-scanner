```python
import os
import time
import yfinance as yf

def download_and_save_to_csv(symbols_list):
    print(f"Starting data download for {len(symbols_list)} symbols using native curl_cffi wrapper...\n")
    counter = 0
    for symbol in symbols_list:
        raw_symbol = symbol.strip().upper()
        counter += 1
        # if os.path.exists(f"{raw_symbol}.csv"):
        #     print(f"{counter}/{len(symbols_list)}. File for {raw_symbol} already exists. Skipping download.")
        #     continue
        
        if not raw_symbol.endswith('.NS') and '.' not in raw_symbol:
            ticker = f"{raw_symbol}.NS"
        else:
            ticker = raw_symbol

        print(f"{counter}/{len(symbols_list)}. Fetchingdata for: {ticker}...")
        
        try:
            df = yf.download(
                tickers=ticker,
                period="max",
                interval="1d",
                auto_adjust=False,  
                threads=False,     
                timeout=15,
                progress=False
            )
            
            if df is not None and not df.empty:
                # --- FIX: FLATTEN MULTI-INDEX COLUMNS IMMEDIATELY ---
                # If the headers look like ('Close', 'HDFCBANK.NS'), drop the ticker name layer
                if isinstance(df.columns, range) or hasattr(df.columns, 'levels') or isinstance(df.columns, yf.multi.MultiIndex) or df.columns.nlevels > 1:
                    df.columns = df.columns.get_level_values(0)
                
                # Ensure columns are completely flat string indexes
                df.columns = [str(col).strip() for col in df.columns]
                
                # Now safely drop holiday rows where 'Close' values don't exist
                if 'Close' in df.columns:
                    df = df.dropna(subset=['Close'])
                else:
                    # Fallback check if the column tracking structure is named slightly differently
                    print(f" -> Warning: 'Close' column not found in headers: {list(df.columns)}")
                
                # Define save location using the user's initial tracking input name
                filename = f"{raw_symbol}.csv"
                
                # Save data frame to disk
                df.to_csv(filename)
                print(f" -> Successfully saved: {os.path.abspath(filename)}")
            else:
                print(f" -> Warning: No data returned for {ticker}")
                
        except Exception as e:
            print(f" -> Error downloading {ticker}: {str(e)}")
            
        # Pacing window to maintain healthy API call streams across 500 records
        time.sleep(0.3)
        print("-" * 50)

```

```python
if __name__ == "__main__":
    # List your symbols here (can be with or without '.NS')
    STOCK_SYMBOLS = symbols#["HDFCBANK", "MAHABANK", "RELIANCE", "TCS"]
    download_and_save_to_csv(STOCK_SYMBOLS)
```
